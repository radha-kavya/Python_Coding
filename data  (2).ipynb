```python
import pandas as pd

df1 = pd.read_csv(r'C:\Users\radhakavya\Downloads\archive (1)\Cars Datasets 2025.csv', encoding='latin1')
df2 = pd.read_csv(r'C:\Users\radhakavya\Downloads\archive (2)\electric_vehicles_spec_2025.csv.csv', encoding='latin1')


df1['Seats'] = pd.to_numeric(df1['Seats'], errors='coerce')


merged_df = pd.merge(df1, df2, left_on='Seats', right_on='seats', how='inner')

merged_df.fillna(0, inplace=True)

print(merged_df )
```

```python
df1['Seats'] = pd.to_numeric(df1['Seats'], errors='coerce')  # convert to numeric
df = pd.merge(df1, df2, left_on='Seats', right_on='seats', how='left')
print(df)
```

```python

combined_df = pd.concat([df1, df2], axis=1)


def deduplicate_columns(columns):
    seen = {}
    new_cols = []
    for col in columns:
        if col not in seen:
            seen[col] = 0
            new_cols.append(col)
        else:
            seen[col] += 1
            new_cols.append(f"{col}.{seen[col]}")
    return new_cols


combined_df.columns = deduplicate_columns(combined_df.columns)
print(combined_df.columns)
```

```python
#3.Grouping and Aggregation

combined_df['HorsePower_clean'] = combined_df['HorsePower'].str.extract(r'(\d+\.?\d*)').astype(float)


grouped_stats = combined_df.groupby('Fuel Types')['HorsePower_clean'].agg(['mean', 'std'])

print(grouped_stats)

```
