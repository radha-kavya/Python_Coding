```python
#8. Handling Missing Data*
import pandas as pd

data = {
    'column_name': [1, None, 3, None, 5]
}

df = pd.DataFrame(data)
df['column_name'] = df['column_name'].interpolate(method='linear')

print(df)
```

```python

import numpy as np
df = pd.DataFrame({'column_name': [1, np.nan, 3, np.nan, 5]})

mask = np.isnan(df['column_name'])
df.loc[mask, 'column_name'] = 0

print(df)
```

```python

col = df['column_name']
median = np.nanmedian(col) 
std = np.nanstd(col)       
mask = (col < (median - 2 * std)) | (col > (median + 2 * std))
df.loc[mask, 'column_name'] = median
print(mask)
```

```python
# *9. Advanced Data Analysis*
import pandas as pd

data = {
    'region': ['North', 'South', 'North', 'South', 'East'],
    'product': ['A', 'B', 'A', 'B', 'A'],
    'sales': [100, 150, 200, 250, 300]
}

df = pd.DataFrame(data)

grouped_df = df.groupby(['region', 'product'])['sales'].sum().reset_index()
print(grouped_df)
```

```python
import pandas as pd

data = {
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [2, 3, 4, 5]
}

df = pd.DataFrame(data)
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'sales': [100, 120, 130, 140, 150, 160, 170, 180, 190, 200]
})

df['rolling_avg'] = df['sales'].rolling(window=7).mean()

# Plot
df['rolling_avg'].plot(title='7-Day Rolling Average of Sales')
plt.xlabel('Index')
plt.ylabel('Sales Rolling Average')
plt.grid(True)
plt.show()
```
